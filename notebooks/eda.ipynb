```python
# Generate comprehensive summary report
summary_report = []

for name, df in dataframes.items():
    print(f"\n{'='*70}")
    print(f"Summary Report: {name.upper()}")
    print(f"{'='*70}")
    
    for col in df.columns:
        missing_count = df[col].isna().sum()
        missing_pct = (missing_count / len(df)) * 100
        unique_count = df[col].nunique()
        data_type = str(df[col].dtype)
        
        report_dict = {
            'Dataset': name,
            'Column': col,
            'DataType': data_type,
            'MissingCount': missing_count,
            'MissingPercent': f'{missing_pct:.2f}%',
            'UniqueValues': unique_count
        }
        
        # Add appropriate statistics based on data type
        if df[col].dtype in [np.int64, np.float64]:
            report_dict['Mean'] = df[col].mean()
            report_dict['Std'] = df[col].std()
            report_dict['Min'] = df[col].min()
            report_dict['Max'] = df[col].max()
        
        summary_report.append(report_dict)
        
        print(f"\n  {col}:")
        print(f"    Data Type: {data_type}")
        print(f"    Missing: {missing_count} ({missing_pct:.2f}%)")
        print(f"    Unique Values: {unique_count}")
        if df[col].dtype in [np.int64, np.float64]:
            print(f"    Mean: {df[col].mean():.4f}")
            print(f"    Std: {df[col].std():.4f}")
            print(f"    Min: {df[col].min():.4f}")
            print(f"    Max: {df[col].max():.4f}")

# Create summary DataFrame
summary_df = pd.DataFrame(summary_report)
print(f"\n\n{'='*70}")
print("COMPREHENSIVE SUMMARY TABLE")
print(f"{'='*70}")
display(summary_df)
```
